```python
import pandas as pd
```

```python
dataset=pd.read_csv('/loan_data_set.csv')
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset['Gender'].fillna(dataset["Gender"].mode()[0],inplace=True)
```

```python
dataset.isnull().sum()
```

```python
dataset['Married'].fillna(dataset['Married'].mode()[0],inplace=True)
```

```python
dataset.isnull().sum()
```

One-Hot encoding

```python
en_data=dataset[["Gender","Married"]]
```

```python
#get dummies
pd.get_dummies(en_data).info()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
one=OneHotEncoder(drop="first")
ar=one.fit_transform(en_data).toarray()
```

```python
pd.DataFrame(ar,columns=["Gender_Male","Married_Yes"])
```
